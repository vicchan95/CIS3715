**Question 5**: What can we conclude by looking at the histogram? Explain in one-two sentences. Figure out what is your preferred number of bins for *MPG* and *Cylinders* and argue why. 

**Answer**: The histograms tell us the frequency that the cars' MPG falls under class interval (bin) 0 to say 10 and then from 10 to 20, etc. This is the same with Cylinders but in smaller numbers from 3 to 4, 4 to 5, etc. From when the parameter of bins was still 20, and changing it to 10, I conclude that the peak frequency of MPG occurs between 13 and 20, and the most frequent Cylinders for cars within the dataset is 4.
Individually, my preferred number of bins for MPG is 7 while for Cylinders it is 5. This is because I played around with the parameters and found that bins=7 normalizes the MPG histogram. By this, I mean the graph starts looking approximately normal distribution with clearly defined borders of where the mode occurs. If I chose a higher number for bins then the graph would have a gap around the 4-10 mark of the x axis which is considered "bad data." Going lower than bin=7 would make the data blend together too much. For the cylinders, I tried aiming for the lowest possible number of bins that doesn't blend the data together and eliminates any gaps. From looking at the raw data this confers with the fact that I can see the mode is 4.

The following code provides statistics for number of cylinders.

```python
t = pd.crosstab(index=df["Cylinders"],  # Make a crosstab
                     columns="count")     # Name the count column
t['percentage']= (t/t.sum())*100
print (t)
```

**Question 6**: Try to learn more about `crosstab` method (by doing Google search) and write a line of code that uses it in a different way on `df` data.

**Answer**:

```python
cross = pd.crosstab(index=df["Origin"],columns=df["Model"])
print(cross)
```

**continuing Answer 6**: I cross tabulated the data between Origin and Model columns to display a table that tells me the count of new models created by each country in the years 1970 to 1982. This allows me to easily count the number of new models created by those three countries (Europe, Japan, and US) each year by summing the columns. Also I can sum the rows up to find the number of new models created by those three countries within that 12 year time period.

Scatterplot between MPG and Weight attributes:

```python
plt.figure(figsize=(5,3))
plt.scatter(df['MPG'],df['Weight']);
```

**Question 10**: Discuss what can we learn from the displayed scatterplot.  

**Answer**: We can learn the correlation between two types of data within our data set. Here, we are comparing MPG with Weight of each car and there seems to be a negative correlation. There are some outliers due to MPG being 0 as the null value for some cars but the trend fits that the more a car weighs the less MPG it tends to have.

We can decide on the size and color of circles to represent some of data properties. Here, the size is proportional to Cylinders and color is related to Horsepower:

```python
plt.figure(figsize=(5,3))
plt.scatter(df['MPG'],df['Weight'],df['Cylinders']*5,df['Horsepower']);
plt.colorbar()
```

**Question 11**: Discuss what additional insight we can get from this scatterplot as compared to the previous scatterplot. Plot another scatterplot by picking a different set of attributes that you think is providing a useful view at the data. Discuss what can be concluded from that plot.

**Answer**: 

```python
plt.figure(figsize=(15,10))
plt.scatter(df['Horsepower'],df['Weight'],df['Acceleration']*6,df['Model'])
plt.title('Horsepower, Weight, Acceleration, and Model comparison')
plt.xlabel('Horsepower',fontsize=15)
plt.ylabel('Weight',fontsize=15)
plt.legend()
plt.colorbar(label='Model')
```

Here, I plotted the horsepower against the weight of each car in the data set. It shows a supposedly positive correlation so I looked up the ratio and found: "A car with a high power to weight ratio will accelerate faster than a automobile with a lower value." Therefore, passing acceleration as an argument representing the size of each plot, I find that plots are bigger in size when they are on the same x-value but higher y-value. Lastly, taking into account the color of the plots, it can be conclude that the horsepower and weight of cars began having a decreasing trend throughout the years from 1970 to 1982. As with horsepower and weight, acceleration also decreased throughout the years.

Pandas `scatter_matrix` method allows us to plot all scatterplots for a data set (it would take a few seconds to display):

```python
pd.scatter_matrix(df, figsize=(20,20));
```

**Question 12**: Explain what are we seeing from this plot and discuss about the insights you obtained from it.

**Answer**: The scatter_matrix function provides us with an array of scatter plots compiled into one big matrix-like shape (hence the name). The function plots each property of the data set against each other, but when the data properties are plotted against itself it produces a histogram instead. This is why the main diagonal of the matrix is a line of histograms.
From this scatter_matrix, I can gain insights through comparison of the scatter plots between one property of the data set and all other properties, the change caused by switching x and y-axises, and what happens when values overlap in a scatter plot. I see that MPG has negative correlation with weight, horsepower, and displacement. I also see that flipping the x- and y-axises of a scatter plot seems to have a mirror effect on the graph, reflecting it along the perfect 1.0 correlation line, but keeping the same sign (+ or -) in correlation. Lastly, using scatter plots for data that is repetitive, such as the cylinders and model, where the data repeats a lot results in the data being grouped into straight lines with lots of overlapping scatter points.
